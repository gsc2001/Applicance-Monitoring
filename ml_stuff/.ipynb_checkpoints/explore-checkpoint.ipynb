```python
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
with open("mod_data.json",'r') as fd:
    df=json.load(fd)
```

```python
len(df)
```

```python
x=np.linspace(1,5,num=5,endpoint=True)
```

```python
x
```

```python
ans_df=dict()
```

```python
matplotlib.rc('figure', figsize=(10, 5))
```

```python
for key,curr_obj in df.items():
    x=list(np.linspace(1, len(curr_obj['data']), num=len(curr_obj['data']), endpoint=True))
    y=curr_obj['data']
    print(key)
    device=key.split()[1]
    if device in ans_df:
        ans_df[device].extend(y)
    else:
        ans_df[device]=y
    fig10 = plt.figure()
    win = fig10.add_subplot(1,1,1)
    fig10 = plt.plot(x, y, lw = 0.1)
    fig10 = plt.xlabel("Time in ms")
    fig10 = plt.ylabel("Current")
    fig10 = plt.title("Time with current in 0.1mA")
    print(pd.DataFrame([abs(t) for t in y]).describe())
    #print(x)
    #break
    fig=plt.plot()
    plt.show()
    print("##############################")
```

```python
for key,curr_obj in df.items():
    x=list(np.linspace(1, len(curr_obj['data']), num=len(curr_obj['data']), endpoint=True))
    y=curr_obj['data']
    print(key)
    fig10 = plt.figure()
    win = fig10.add_subplot(1,1,1)
    fig10 = plt.scatter(x, y, s = 0.1,label=key)

    fig10 = plt.xlabel("Time in ms")
    fig10 = plt.ylabel("Current")
    fig10 = plt.title("Time with current in mA")
    print(pd.DataFrame([abs(t) for t in y]).describe())
    #print(x)
    #break
    fig=plt.plot()
    plt.show()
    print("##############################")
```

```python
x_int=600
y_int=10000
for key,curr_obj in df.items():
    x=list(np.linspace(1, len(curr_obj['data']), num=len(curr_obj['data']), endpoint=True))
    y=curr_obj['data']
    print(key)
    #fig10 = plt.figure()
    #win = fig10.add_subplot(1,1,1)
    plt.scatter(x, y, s = 0.5,label=key)
    x_line=list(np.arange(0,np.max(x),x_int))
    y_line=list(np.arange(-np.max(y),np.max(y),y_int))
    
    for xc in x_line:
        plt.axvline(x=xc, color='r', linestyle='--')
    
    for yc in y_line:
        plt.axhline(y=yc, color='g', linestyle='--')
    
    

    print(pd.DataFrame([abs(t) for t in y]).describe())
    fig=plt.plot()
    plt.show()
    #print(x)
    #break
    
    print("##############################")
```
