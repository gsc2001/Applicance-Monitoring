```python
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
with open("mod_data.json",'r') as fd:
    df=json.load(fd)
```

```python
matplotlib.rc('figure', figsize=(10, 5))
```

```python
for key,curr_obj in df.items():
    x=list(np.linspace(1, len(curr_obj['data']), num=len(curr_obj['data']), endpoint=True))
    y=curr_obj['data']
    print(key)
    fig10 = plt.figure()
    win = fig10.add_subplot(1,1,1)
    fig10 = plt.plot(x, y, lw = 0.1)
    fig10 = plt.xlabel("Time in ms")
    fig10 = plt.ylabel("Current")
    fig10 = plt.title("Time with current in mA")
    print(pd.DataFrame([abs(t) for t in y]).describe())
    #print(x)
    #break
    fig=plt.plot()
    plt.show()
    print("##############################")
```
