```python
import json
from copy import deepcopy
```

```python
with open("data.json",'r') as fd:
    df=json.load(fd)
```

```python
df[360]
```

```python
df=list(filter(lambda x:('iteration' in x) and (x['device']!='T'), df))
```

```python
df=list(filter(lambda x:(len(x['data'])%6000)==0, df))
```

```python
len(df)
```

```python
df=list(sorted(df, key=lambda x:x['iteration']))
```

```python
len(df)
```

```python
final_df={}
for curr_arr in df:
    curr_key=curr_arr['sender']+" "+curr_arr['device']+" "+str(curr_arr['version'])
    if curr_key not in final_df:
        final_df[curr_key]=deepcopy(curr_arr)
    else:
        final_df[curr_key]['data'].extend(curr_arr['data'])
    final_df[curr_key]['total_points']=len(final_df[curr_key]['data'])
            
```

```python
len(final_df)
```

```python
print(*list(final_df.keys()),sep='\n')
```

```python
for key, val in final_df.items():
    print(f"{key} -> {len(val['data'])}")
```

```python
with open("mod_data.json",'w') as fd:
    json.dump(final_df, fd, indent=4)
```
