```python
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
with open("mod_data.json",'r') as fd:
    df=json.load(fd)
```

```python
interval_length=300
FILE_NAME=f"{interval_length}_data.json"
```

```python
dataset=dict()
```

```python
from copy import deepcopy
```

```python
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)
```

```python
for curr_key, curr_val in df.items():
    tot_len=len(curr_val['data'])
    assert(tot_len%interval_length==0)
    array_parts=np.array_split(curr_val['data'], tot_len/interval_length)
    array_parts=[list(x) for x in array_parts]
    #print(len(array_parts))
    dev=curr_val['device']
    if dev not in dataset:
        dataset[dev]=[]
    dataset[dev].extend(deepcopy(array_parts))
```

```python
with open(FILE_NAME,'w') as fd:
    json.dump(dataset, fd, indent=4,cls=NpEncoder)
```
